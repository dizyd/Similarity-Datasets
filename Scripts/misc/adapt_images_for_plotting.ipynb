This script:

1. Reduces the size of each image by 50%
2. Converts white pixels to transparent 
3. Strip meta data to reduce file size
4. Compresses the image when saving to further reduce file size

```python
# Import packages

from PIL import Image
import os
```

```python
# Define image processing  function
def process_image(input_path, output_path, scale = 0.5, white_threshold = 240, domain):
    
    # Open image and convert to RGBA
    img = Image.open(input_path).convert("RGBA")
    width, height = img.size

    # Resize image
    img = img.resize((int(width*scale), int(height*scale)),resample = Image.LANCZOS)

    # Make white pixels transparent
    datas    = img.getdata()
    new_data = [] 

    for item in datas:
        r, g, b, a = item
        if r > white_threshold and g > white_threshold and b > white_threshold:
            new_data.append((255,255,255,0)) # transparent
        else: 
            new_data.append((r, g, b, a))
    
    img.putdata(new_data)

    # Create new image to ensure metadata i stripped
    clean_img = Image.new("RGBA",img.size)
    clean_img.putdata(img.getdata())

    # Save image with compression
    clean_img.save(output_path,optimize=True)
```

```python
# Countries
input_folder  = "../Materials/images/Countries" 
output_folder = "../Materials/images/Reduce Quality/Countries" 

for filename in os.listdir(input_folder):
    if filename.lower().endswith((".png",".jpg",".jpeg")):
        input_file = os.path.join(input_folder, filename)
        ouput_file = os.path.join(output_folder, os.path.splitext(filename)[0] + ".png")
        process_image(input_file,ouput_file)
```
